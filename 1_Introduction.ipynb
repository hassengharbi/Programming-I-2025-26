# Functions in Pyhton

## Creating a Function without parameter nor returning value

```python
def AreaCircle():
    radius = 10
    pi=3.14
    area=pi * (radius**2)
    print(area)
```

# Put the Function in Action by calling it 

```python
AreaCircle()
```

## Creating a Function with parameter but without returning value

```python
def AreaCircle1(radius):
    pi=3.14
    area=pi * (radius**2)
    print(area)
```

```python
AreaCircle1(20)
AreaCircle1(21)
AreaCircle1(22)


```

## Creating a Function with parameter and returning value

```python
def AreaCircle2(radius):
    pi=3.14
    area=pi * (radius**2)
    return area
```

```python
AreaCircle2(3)
```

## The returning value of a Function could be assigned to an other variable

```python
result=AreaCircle2(12)
print(result)
```

# Scope of Variables

## Local Variables vs Global Variables

### Example 1 

```python
c=0
def counter():
    return c

print("First call",counter())
print("Second call",counter())
print("Third call",counter())
```

### Example 2 

```python
c=0
def counter2():
    c=c+1
    return c

print("First call",counter2())
print("Second call",counter2())
print("Third call",counter2())
```

### Example 3

```python
myVariable = 7
myParam = 20

def func1(myParam):
    myVariable = 20
    print(myParam)
    print("myVaribale inside func1",myVariable)

func1(5)
print(myVariable)
```

### Use of global variable inside a function script with the keyword : 'global'

```python
cpt = 0
def counter():
    global cpt
    cpt=cpt+1
    return cpt

print("First call",counter())
print("second call",counter())
print("Third call",counter())
```
