```python
# create a function that will create a file in append mode 
# the name of the file is given as an argument of our function
def creatf(fname):
    f=open(f"{fname}.txt",'a')
    f.close()
    print(f"File {fname}.txt is created")   
```

```python
creatf('testfile2')
```

```python
# Write a Function that rewrites file with user input text and file name
# given as arguments
def writef(text,fname):
    try:
        f=open(f'{fname}.txt', 'w')
        f.write(text)
        f.close
    except FileNotFoundError :
        print('please take care of all errors')
        
```

```python
writef('Hello !', 'hello')
```

```python
# Write a Function that will add a text to a file without rewriting it
# text and file name are given as arguments 
def addf(text,fname):
    try:
        f=open(f'{fname}.txt','a')
        f.write(text)
        f.close
    except FileNotFoundError:
        print(f'File {fname}.txt does not exist')
```

```python
addf('Hellooo','Hellloo')
```
