## 2- Write a Python function named newPoint(x,y,c) to create new point as a dictionary that will be returned. More, this function display the message : "New Point is created  (x=... | y=... | color=...)"

```python
# write your code here
def newPoint(x,y,c):
    p = {'x':x, 'y':y, 'color': c}
    msg = f"New Point is created (x={x} | y = {y} | color = {c})"
    print(msg)
    return p
```

```python
#Example of use of the function newPoint
newp=newPoint(0,0,'red')
print(newp)
```

## 3- Write a Python function named addPoint to add a point to listPoint. The point is given as a parameter to the function addPoint. More, this function display the message : " Point (x=... | y=... | color=...) added to list of Points"

```python
# write your code here

def addPoint(p):
    #p is a dictionary representing a point
    listPoint.append(p)
    msg = f"Point(x={p['x']} | y = {p['y']} | color = {p['c']})"
    print(msg)
```

## 4- Write a Python function named movePoint(index,dx,dy) to move the point at index position into listPoint. The move is done by dx et dy. The first point has index 1. More, this function display the message : " After move Point[index] has (x=... | y=... | color=...)". if the index does not exist in listPoint then dispaly "No such index"

```python
# write your code here
def movePoint(index,dx,dy):
    if index>=0 and index<=len(listPoint):
        listPoint[index-1]['x'] += dx
        listPoint[index-1]['y'] += dy
        msg = f"Point[{index}] has now (x={listPoint[index-1]['x']} | y = {listPoint[index-1]['y']} | color = {listPoint[index-1]['color']})"
        print(msg)
    else:
        print('No such Index')
```

## 5- Write a Python function named colorPoint(index,newcolor) to change the color of the point at index position to newcolor into listPoint. The first point has index 1. ## 6- Write a Python function named distPoint(point1,point2) to calculate the distance between point1 and point2 and return it. import math module

```python
# write your code here
def distPoint(point1, point2):
    import math
    distance = math.sqrt((point1['x']-point2['x'])**2 + (point1['y']-point2['y'])**2)
    return distance
```

```python
p1={'x':1 , 'y':1, 'color':'red'}
p2={'x':2 , 'y':2, 'color':'red'}
d= distPoint(p1,p2)
print(d)
```
